```python
def load_format_data():
    df = pd.read_csv('input/NCVACompositeMetrics.csv')
    #df = pd.read_csv('input/CompositeMetrics.csv')
    df['st_fips'] = df['st_fips'].astype('str')
    df['st_fips'] = df['st_fips'].str.zfill(2)
    df['co_fips'] = df['co_fips'].astype('str')
    df['co_fips'] = df['co_fips'].str.zfill(5)
    return df

df = load_format_data()
    
df
```

Create filter for just NC data; ST_FIPS == 37 and then select only the columns we will be working with; ['CO_FIPS','DATE','CASES','DEATHS']

```python
#SIMULATE CAN ASSIGN VALUES TO THESE PARAMETERS
parameters = {'state_fips': ['51']}
```

```python
def filter_for_state(df, FIPS_STRING_LIST):
    state_filter = df['st_fips'].isin(FIPS_STRING_LIST)
    df = df[state_filter]
    df = df[['st_fips', 'state', 'co_fips','name','date','cases', 'deaths']]
    # make df compatible with the NYT file that Donal uses as input      
    df.columns = df.columns.str.lower()
    df.rename(columns={'co_fips': 'fips', 'name': 'county'}, inplace=True)
    return df

#statelist = ['37','51']
#statelist = STATE_LOOKUP_DF['state_fips'].tolist()

output_df = filter_for_state(df, parameters['state_fips'])

output_df
```
